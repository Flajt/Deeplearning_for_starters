A Neural Network is currently the best (not **the best**, but the best currently). <br>
It uses a math and programming to create a brain like programm to solve diffrent kinds of problems. <br>
For each probelm is an other Network to use. <br>
The image above shows you a modell of a Network that consists of 4 Layers:
1. The `Input Layer`: Here data (like images) in form of integers (realy numbers) or floats (numbers with commata) will be inserted.
2. The second layer is called `Hidden Layer` and because we have two of them it's the first hidden layer. Here the data of all other neurons (small balls) will be summed up, a `Bias` will be applied (a simple number) and then all this went through an activation function (for example a sigmoid) in the next layer.
3. Second hidden Layer, just the same as before.
4. `Ouput Layer`, does the same as the other hidden layers but returns the values.
<br>
The lines are the connection of one neuron to the other, but there is aswell a weight matrix, wich "stores" the knowlege of the Network. The data that went through to the next neuron will be multiplied by that weight sum, before it will be summed up in the next neuron.

The error will then be propaged backwarts to adjust the weights a little bit, this process will be repeated 10000 times over to get the best weights.
